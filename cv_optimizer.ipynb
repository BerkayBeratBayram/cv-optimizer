## HÃ¼cre 1: KÃ¼tÃ¼phaneleri Ä°Ã§e Aktar

```python
def read_cv(file_path):
    """
    CV'yi oku (PDF veya TXT)
    """
    print(f"ğŸ“„ CV okunuyor: {file_path}")
    
    if file_path.endswith('.pdf'):
        # PDF okuma
        reader = PdfReader(file_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"
    
    elif file_path.endswith('.txt'):
        # Text dosyasÄ±
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
    
    else:
        raise ValueError("Sadece PDF veya TXT desteklenir!")
    
    print(f"âœ… {len(text)} karakter okundu")
    print(f"ğŸ“Š Kelime sayÄ±sÄ±: {len(text.split())}")
    
    return text

print("âœ… Fonksiyon hazÄ±r!")
```

## HÃ¼cre 3: Keyword Extraction (Anahtar Kelime Ã‡Ä±karma)

```python
def extract_keywords(text):
    """
    Metinden beceri ve teknoloji kelimelerini Ã§Ä±kar
    """
    # YaygÄ±n teknoloji ve beceri kelimeleri (basit pattern matching)
    keywords = set()
    
    # Programlama dilleri
    languages = ['python', 'javascript', 'java', 'c++', 'c#', 'ruby', 'go', 'rust', 
                 'php', 'swift', 'kotlin', 'typescript', 'scala', 'r']
    
    # Framework'ler
    frameworks = ['django', 'flask', 'fastapi', 'react', 'vue', 'angular', 
                  'express', 'spring', 'node.js', 'next.js', 'rails']
    
    # VeritabanlarÄ±
    databases = ['postgresql', 'mysql', 'mongodb', 'redis', 'cassandra', 
                 'oracle', 'sql server', 'dynamodb', 'elasticsearch']
    
    # Cloud & DevOps
    cloud = ['aws', 'azure', 'gcp', 'docker', 'kubernetes', 'jenkins', 
             'terraform', 'ansible', 'ci/cd', 'gitlab']
    
    # DiÄŸer teknolojiler
    other = ['git', 'linux', 'rest api', 'graphql', 'microservices', 
             'agile', 'scrum', 'tdd', 'machine learning', 'deep learning',
             'celery', 'rabbitmq', 'kafka', 'spark']
    
    all_keywords = languages + frameworks + databases + cloud + other
    
    text_lower = text.lower()
    
    for keyword in all_keywords:
        if keyword in text_lower:
            keywords.add(keyword)
    
    return sorted(list(keywords))

# Test edelim
cv_text_content = read_cv("data/ornek_cv.txt")
cv_keywords = extract_keywords(cv_text_content)

print("ğŸ”‘ CV'deki anahtar kelimeler:")
print(", ".join(cv_keywords))
```

## HÃ¼cre 4: ChromaDB Embedding FonksiyonlarÄ±

```python
def create_cv_embeddings(cv_path, job_path):
    """
    ChromaDB ile CV ve iÅŸ ilanÄ± embedding'leri oluÅŸtur
    """
    print("ğŸ”® ChromaDB ile embedding oluÅŸturuluyor...")
    
    # ChromaDB client oluÅŸtur
    client = chromadb.Client()
    
    # Collection oluÅŸtur (varsa sil)
    try:
        client.delete_collection("cv_job_matcher")
    except:
        pass
    
    collection = client.create_collection(
        name="cv_job_matcher",
        metadata={"description": "CV ve iÅŸ ilanÄ± eÅŸleÅŸtirme"}
    )
    
    # CV'yi oku ve parÃ§alara bÃ¶l
    cv_text = read_cv(cv_path)
    job_text = read_cv(job_path)
    
    # Text splitter ile parÃ§ala
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50,
        separators=["\n\n", "\n", ". ", " ", ""]
    )
    
    cv_chunks = text_splitter.split_text(cv_text)
    job_chunks = text_splitter.split_text(job_text)
    
    print(f"ğŸ“¦ CV {len(cv_chunks)} parÃ§aya bÃ¶lÃ¼ndÃ¼")
    print(f"ğŸ“¦ Ä°ÅŸ ilanÄ± {len(job_chunks)} parÃ§aya bÃ¶lÃ¼ndÃ¼")
    
    # CV parÃ§alarÄ±nÄ± ekle
    for i, chunk in enumerate(cv_chunks):
        collection.add(
            documents=[chunk],
            metadatas=[{"source": "cv", "chunk_id": i}],
            ids=[f"cv_{i}"]
        )
    
    # Ä°ÅŸ ilanÄ± parÃ§alarÄ±nÄ± ekle
    for i, chunk in enumerate(job_chunks):
        collection.add(
            documents=[chunk],
            metadatas=[{"source": "job", "chunk_id": i}],
            ids=[f"job_{i}"]
        )
    
    print("âœ… Embedding'ler oluÅŸturuldu!")
    return collection

def find_matching_sections(collection, query, n_results=5):
    """
    Query'ye gÃ¶re en alakalÄ± bÃ¶lÃ¼mleri bul
    """
    results = collection.query(
        query_texts=[query],
        n_results=n_results,
        include=["documents", "metadatas", "distances"]
    )
    
    return results

def semantic_cv_analysis(cv_path, job_path):
    """
    Semantic analiz ile CV ve iÅŸ ilanÄ± karÅŸÄ±laÅŸtÄ±rmasÄ±
    """
    print("=" * 70)
    print("ğŸ”® SEMANTÄ°K ANALÄ°Z (CHROMADB)")
    print("=" * 70)
    
    # Embedding'leri oluÅŸtur
    collection = create_cv_embeddings(cv_path, job_path)
    
    # Ä°ÅŸ ilanÄ±ndan Ã¶nemli gereksinimleri Ã§Ä±kar
    job_text = read_cv(job_path)
    
    # Ä°ÅŸ ilanÄ±nÄ±n tamamÄ±nÄ± query olarak kullan
    print("\nğŸ“Š Ä°ÅŸ ilanÄ±na en uygun CV bÃ¶lÃ¼mleri:")
    print("=" * 70)
    
    results = find_matching_sections(collection, job_text, n_results=5)
    
    for i, (doc, metadata, distance) in enumerate(zip(
        results['documents'][0],
        results['metadatas'][0],
        results['distances'][0]
    )):
        similarity = 1 - distance  # Distance'Ä± similarity'e Ã§evir
        source = "ğŸ“„ CV" if metadata['source'] == 'cv' else "ğŸ“‹ Ä°lan"
        
        print(f"\n{i+1}. {source} - Benzerlik: {similarity:.2%}")
        print(f"   {doc[:200]}...")
    
    # CV'den spesifik becerileri sorgula
    print("\n" + "=" * 70)
    print("ğŸ¯ Spesifik beceri aramalarÄ±:")
    print("=" * 70)
    
    skill_queries = [
        "Python ve Django deneyimi",
        "VeritabanÄ± yÃ¶netimi ve SQL",
        "Cloud platformlarÄ± AWS GCP",
        "TakÄ±m Ã§alÄ±ÅŸmasÄ± ve Agile",
        "Machine Learning ve AI"
    ]
    
    for query in skill_queries:
        print(f"\nğŸ” Arama: {query}")
        results = find_matching_sections(collection, query, n_results=2)
        
        for doc, metadata in zip(results['documents'][0], results['metadatas'][0]):
            source = "âœ… CV'de var" if metadata['source'] == 'cv' else "ğŸ“‹ Ä°landa isteniyor"
            print(f"   {source}: {doc[:150]}...")
    
    return collection

# Test edelim
collection = semantic_cv_analysis("data/ornek_cv.txt", "data/ornek_ilan.txt")
```

## HÃ¼cre 7: Basit Keyword KarÅŸÄ±laÅŸtÄ±rma

```python
def compare_cv_job(cv_path, job_path):
    """
    CV ile iÅŸ ilanÄ±nÄ± karÅŸÄ±laÅŸtÄ±r
    """
    print("=" * 70)
    print("ğŸ“Š CV VE Ä°Å Ä°LANI KARÅILAÅTIRMA")
    print("=" * 70)
    
    # CV'yi oku
    cv_text = read_cv(cv_path)
    cv_keywords = set(extract_keywords(cv_text))
    
    # Ä°ÅŸ ilanÄ±nÄ± oku
    job_text = read_cv(job_path)
    job_keywords = set(extract_keywords(job_text))
    
    # KarÅŸÄ±laÅŸtÄ±rma
    matched = cv_keywords & job_keywords  # KesiÅŸim
    missing = job_keywords - cv_keywords  # CV'de yok
    extra = cv_keywords - job_keywords    # Ä°landa yok ama CV'de var
    
    match_percentage = (len(matched) / len(job_keywords) * 100) if job_keywords else 0
    
    print(f"\nğŸ“ˆ EÅŸleÅŸme OranÄ±: {match_percentage:.1f}%")
    print(f"âœ… EÅŸleÅŸen: {len(matched)}")
    print(f"âŒ Eksik: {len(missing)}")
    print(f"â• Fazladan: {len(extra)}")
    
    print("\n" + "=" * 70)
    print("âœ… CV'DE OLAN VE Ä°Å Ä°LANINDA ARANAN BECERÄ°LER:")
    print("=" * 70)
    for keyword in sorted(matched):
        print(f"  â€¢ {keyword}")
    
    print("\n" + "=" * 70)
    print("âŒ CV'DE OLMAYAN AMA Ä°Å Ä°LANINDA Ä°STENEN BECERÄ°LER:")
    print("=" * 70)
    for keyword in sorted(missing):
        print(f"  â€¢ {keyword}")
    
    if extra:
        print("\n" + "=" * 70)
        print("â• CV'DE OLAN AMA Ä°Å Ä°LANINDA BELÄ°RTÄ°LMEYEN BECERÄ°LER:")
        print("=" * 70)
        for keyword in sorted(extra):
            print(f"  â€¢ {keyword}")
    
    return {
        'cv_keywords': cv_keywords,
        'job_keywords': job_keywords,
        'matched': matched,
        'missing': missing,
        'extra': extra,
        'match_percentage': match_percentage
    }

# Test
result = compare_cv_job("data/ornek_cv.txt", "data/ornek_ilan.txt")
```

## HÃ¼cre 8: ChromaDB + AI GeliÅŸmiÅŸ Analiz

```python
def ai_with_embeddings(cv_path, job_path):
    """
    ChromaDB embedding + AI ile geliÅŸmiÅŸ analiz
    """
    print("ğŸš€ ChromaDB + AI ile derin analiz yapÄ±lÄ±yor...")
    
    # Embedding'leri oluÅŸtur
    collection = create_cv_embeddings(cv_path, job_path)
    
    # Ä°ÅŸ ilanÄ±nÄ± oku
    job_text = read_cv(job_path)
    
    # Ä°ÅŸ ilanÄ±na gÃ¶re CV'den en alakalÄ± bÃ¶lÃ¼mleri bul
    print("\nğŸ” Ä°ÅŸ ilanÄ±na en alakalÄ± CV bÃ¶lÃ¼mleri bulunuyor...")
    relevant_cv_parts = find_matching_sections(collection, job_text, n_results=3)
    
    # En alakalÄ± CV parÃ§alarÄ±nÄ± birleÅŸtir
    cv_relevant = "\n\n".join([
        doc for doc, meta in zip(
            relevant_cv_parts['documents'][0],
            relevant_cv_parts['metadatas'][0]
        ) if meta['source'] == 'cv'
    ])
    
    # Tam CV'yi de oku
    cv_text = read_cv(cv_path)
    
    # AI'ya daha odaklÄ± prompt ver
    prompt = f"""IMPORTANT: Respond ONLY in Turkish. No English words.

Sen bir TÃ¼rk CV uzmanÄ±sÄ±n. Semantic analiz ile iÅŸ ilanÄ±na EN UYGUN CV bÃ¶lÃ¼mlerini buldum.

Ä°Å Ä°LANINDAKÄ° BAÅLICA GEREKSÄ°NÄ°MLER:
{job_text[:1000]}

CV'DEN Ä°Å Ä°LANINA EN ALAKALI BÃ–LÃœMLER:
{cv_relevant[:1500]}

TAM CV:
{cv_text[:1000]}

GÃ–REVÄ°N (SADECE TÃœRKÃ‡E):
1. AdayÄ±n hangi yÃ¶nleri iÅŸ ilanÄ±yla GÃœÃ‡LÃœ ÅEKÄ°LDE eÅŸleÅŸiyor? (3 madde)
2. Hangi kritik beceriler EKSÄ°K? (3 madde)
3. CV'de hangi deneyimler Ã–NE Ã‡IKARILMALI? (3 Ã¶neri)
4. Ä°ÅŸ ilanÄ±na gÃ¶re CV'ye MUTLAKA eklenecek 3 ÅŸey
5. Genel deÄŸerlendirme (1-10 puan)

CEVAP (TAMAMEN TÃœRKÃ‡E):"""
    
    try:
        response = ollama.generate(
            model='llama3.2',
            prompt=prompt
        )
        
        print("\n" + "=" * 70)
        print("ğŸ¯ CHROMADB + AI DETAYLI ANALÄ°Z:")
        print("=" * 70)
        print(response['response'])
        
        return response['response']
        
    except Exception as e:
        print(f"âŒ Hata: {e}")
        return None

# Test
result = ai_with_embeddings("data/ornek_cv.txt", "data/ornek_ilan.txt")
```
